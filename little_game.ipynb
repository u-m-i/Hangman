```python
""" First models """


from datetime import datetime
from time import sleep

class Character:

  """ Model for any character
  """

  def __init__(self, name : str, species: str):

    self.name = name
    self.species = species
    self.created = datetime.now()
    self.health = 1200
    self.level = 0
    self.tool = 0
    

  def take_weapon(self, arm):
    self.weapon = arm
    print(f"Has tomado : {self.weapon.name} ")
  

  def take_tool(self, tool):
    self.tool = tool
    print(f"Has tomado : {self.tool.name}")


  def attack(self, enemy):
    """ retrives an attack to the elected target
    """
    print("Elige con qué lo quieres agredir\n")
    weapon = input(f"Puño[P], {self.weapon.name}[C]: ").upper()
    if weapon == "P":
        enemy.health -= 150
    elif weapon == "C":
        enemy.health -= self.weapon.damage
    else:
        raise Exception("Tomate esto enserio pedazo de ser hijo de puta, andá a recargarte en la concha de tu hermana")

    return print(f"Has reducido la vida del hijo de puta a {enemy.health}")
    


class Tool:

  """ Model for the objects
  """


  def __init__(self, name , carrier):
    self.name = name
    self.kind = "Common"
    self.carrier = carrier
    self.damage = 500
    

  def throw_it(self):
    """ Only go to trash
    """
    self.carrier = None
    return print("Has desechado tu arma")


  def give_away(self, friend):
    """ Give to somebody else
    """
    self.carrier = friend
    return print(f"Le has sedido tu arma a {friend.name}")


class Knife(Tool):

  """ Basic weapon
  """

  def __init__(self, name, carrier):

    """ 
    """
    return super().__init__(name, carrier)



class Book(Tool):

  """ Tool for the character, helps improving the weapons
  """

  def __init__(self, name, carrier):
      """
      """
      return super().__init__(name, carrier)
    

  def improve(self, arm):
    arm.damage += 150
    return print(f"Se te ha mejorado {arm.name}, ahora tiene 150 más de daño\n")


class Student(Character):

  """ Character kind
  """

  def __init__(self, name, species):

    return super().__init__(name, species)
    self.health = 500
      
 
  def cure(self):
    """ Restart your life
    """
    self.health += 200
        

class DwarfTroll(Character):

  def __init__(self, name, species):
      return super().__init__(name, species)

  def attack(self, enemy):
      enemy.health -= self.weapon.damage


  def take_weapon(self, arm):
    self.weapon = arm

class UltraSword(Tool):

    def __init__(self, name, carrier):
        super().__init__(name, carrier)
        self.damage = 4000000
    




def run():

    troll_malparido = DwarfTroll("Troll enano", "Ser miserable")
    daga = Knife("Daga", troll_malparido)
    troll_malparido.take_weapon(daga)

    albeiro = DwarfTroll("albeiro", "Traicionero hijo de puta")
    mega_espada = UltraSword("PIPI", albeiro)
    albeiro.take_weapon(mega_espada)
    
    print("                         ", "$%"*90, end="\n")
    sleep(3)
    print("Bienvenido a el trabajo que no quería hacer, pero me tocó")
    sleep(3)
    input("Presiona ENTER tecla para continuar. ")
    sleep(1)

    name = input("Dame el nombre de tu personaje: ")
    character = Student(name, "Homeless")
    sleep(2)

    print(f"Perfecto ahora eres {name}, un estudiante vagabundo")
    sleep(2)

    print("Oh mira es un Cuchillo y un libro puedes tomarlos\n")
    sleep(1)
    takeit = input("Presiona T para tomarlos: ").upper()
    if takeit != "T" : raise Exception("Te dije T, no estés mamando y vuelve a empezar")
    sleep(1)
    cuchillo = Knife("Cuchillo", character)
    libro = Book("Morelomicón", character)

    character.take_weapon(cuchillo)
    character.take_tool(libro)
    sleep(3)

    print("Okay,amigo es hora de estrenar nuestro Cuchillo, ya que se aproxima un troll enano con daga")
    sleep(6)
    print("·················································· Has entrado en una pelea ······················································")
    sleep(3)
    print("El Troll ataca primero\n")
    troll_malparido.attack(character)
    sleep(3)


    print(f"Tu actual vida: {character.health} ")
    sleep(2)
    print("Ahora es tu turno")
    sleep(2)
    action = input("Mejorar tu arma [M], atacar de una vez [A]: ").upper()

    sleep(3)
    if action == "M" :
        character.tool.improve(cuchillo)
        sleep(3)
        character.attack(troll_malparido)
    elif action == "A":
        character.attack(troll_malparido)
        sleep(3)
    else:
        raise Exception("Pongase serio malparido")
    sleep(3)
    print("Muy bien, lo has asesinado y humillado\n")

    sleep(3)
    print("Ahora vamos a curarnos")
    sleep(3)
    character.cure()
    print(f"Tu actual vida: {character.health} \n")
    sleep(3)

    print("Mira, un niño vagabundo llamdo albeiro, será mejor darle un cuchillo para que se defienda.")
    input("Presiona ENTER para regarselo ")
    character.weapon.give_away(albeiro)
    sleep(3)
    print("Has hecho una buena acción, ahora tienes 200 más de vida")
    character.health += 200
    print(f"Tu Salud actual {character.health}")
    sleep(5)
    
    print("Oh no parece que albeiro te quiere atacar")
    sleep(4)
    print("·················································· Has entrado en una pelea ······················································")
    sleep(3)
    print("Albeiro ataca primero\n")
    albeiro.attack(character)
    print(f"Tu Salud actual : {character.health}\n")
    sleep(2)
    print("·················································· Has muerto ······················································")
    sleep(3)

    print("Bueno, hasta aquí llega nuestra aventura")
    print(f"Ahora bien, me gustó conocerter {name} nos veremos en otra ocasión")


if __name__ == "__main__":
    run()
    

```
